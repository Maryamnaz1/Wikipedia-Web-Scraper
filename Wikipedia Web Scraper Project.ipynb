```python
def check_title():
    URL = 'https://en.wikipedia.org/wiki/Main_Page'

    page = requests.get(URL)
    Soup1 = BeautifulSoup(page.text,'html')
    articles = Soup1.find(id = 'articlecount').get_text() 
    title = Soup1.find(title = '1867 United States Senate election in Pennsylvania').get_text()
    title = title.strip()
    articles = articles.strip()
    import datetime
    today = datetime.date.today()
    import csv
    header = ['title','articles','Date']
    data = [title,articles,today]

    with open('WikipediaWebScraperProject.csv','w', newline = '',encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data) 
```

```python
while(True):
    check_title()
    time.sleep(86400)
    
```
